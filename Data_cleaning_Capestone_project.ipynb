```python
#we are going to look at the data and find all the unique genres
uniqueGenres = set()
for i in data['genres']:  #i am looping through all the rows
  genlist = ast.literal_eval(i) #convert the string to list of dict
  for j in genlist: #loop through the list of dictionaries
    uniqueGenres.add(j['name']) #add the name of genres to my unique genres
```

```python
print(uniqueGenres)
len(uniqueGenres)
```

```python
rows = []
genres = list(uniqueGenres)
for i in data['genres']:
  genlist = ast.literal_eval(i)
  movieGenres = [x['name'] for x in genlist]
  #print(movieGenres)
  #store a list of 1s and 0s for one movie
  row = []
  for j in genres:
    row.append(int(j in movieGenres))
  #print(row)
  rows.append(row)

```

```python
moviegenres = ['Action', 'Western', 'Drama']
uniqueGenres = ['Action', 'War', 'Documentary', 'Western', 'Music', 'Drama']
row = []
for j in uniqueGenres:
  row.append(int(j in moviegenres))
print(row)
```

```python
genre = pd.DataFrame(rows,columns=genres)
genre
```
